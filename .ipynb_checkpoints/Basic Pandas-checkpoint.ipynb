# Practice using Python Lists

```python
import csv
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import norm
import matplotlib.mlab as mlab

from sklearn.preprocessing import MinMaxScaler
from IPython.display import display, HTML
```

```python
# Function to read in a csv file, rearrange into column wise data format
def CSVdataRead(filename):
    
# Read data using csv reader
    #filename = 'raw_data.csv'
    f = open(filename)
    reader = csv.reader(f)

# append each row of filename into a python list
    rowWise = []

    for i in reader:
        rowWise.append(i)

# remap data from each entry being a row, to each being a column
    colWise = []

    j = 0
    for i in range(len(rowWise[0])):       # for every column (found by specifying length of first item in rowWise)
        temp = []                          # Initialize list before going into loop
        for j in range(len(rowWise)):      # Each line in temp, append with j'th row, i'th column
            temp.append(rowWise[j][i])
        colWise.append(temp)               # append colWise list with temp (essentially a column of data)
    return colWise
filename = 'raw_data.csv'        
lists = CSVdataRead(filename)

```

# Bin data with labels

```python
a = pd.DataFrame({'data': np.array([1,2,3,13,8,20,13,22,50,60,75,45,30,22,11,12,13,14,15,16,71,18])})

b = pd.cut(a['data'],6, retbins=True)
c = b[1]
#print(b[0][0])
#print(c)
num_bins = len(c)-1
bin_dict = {}
binny ={}
#print(b[0])

#for i in range(len(b[0])):
    #print(b[0][i])

for i in range(num_bins):
    
    #print(i,i+1)
    
    bin_dict[f'bin_{i+1}'] = b[1][i],b[1][i+1]
    binny[f'bin_{i+1}'] = []

print(binny)
a['data'][0]
for i in range(len(a['data'])):
    for j in range(num_bins):
        d = bin_dict[f'bin_{j+1}']
        if a['data'][i] >= d[0] and a['data'][i] <= d[1]:
            binny[f'bin_{j+1}'].append(i)
            
        
print(bin_dict)
print(binny)
print(a['data'][10])
```

# Normalize data with MinMaxScaler 

```python
# Import data with pandas
train = pd.read_csv('train.csv')
train.isnull()
train = train.dropna(how='any')

P_Fare = train['Fare']
P_Fare_np=np.array(P_Fare)
sc = MinMaxScaler(feature_range=(0,1))
tf = sc.fit_transform(P_Fare.values.reshape(-1,1))

mx = max(P_Fare_np)
mn = min(P_Fare_np)

norm = []

for i in range(len(P_Fare_np)):
    temp = (P_Fare_np[i] - mn)/(mx-mn)

    norm.append(temp)


plt.plot(tf)
plt.plot(norm)
plt.xlabel('Array Entry')
plt.ylabel('MinMaxScaler - Norm')

comp = []
for i in range(len(tf)):

    comp.append(tf[i]-norm[i])
    
compPD = pd.DataFrame(data=comp)
print('See, MinMaxScaler is just normalizing data from 0 to 1')
compPD.head(5)
```
