# Practice using Python Lists

```python
import csv
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Function to read in a csv file, rearrange into column wise data format
def CSVdataRead(filename):
    
# Read data using csv reader
    #filename = 'raw_data.csv'
    f = open(filename)
    reader = csv.reader(f)

# append each row of filename into a python list
    rowWise = []

    for i in reader:
        rowWise.append(i)

# remap data from each entry being a row, to each being a column
    colWise = []

    j = 0
    for i in range(len(rowWise[0])):       # for every column (found by specifying length of first item in rowWise)
        temp = []                          # Initialize list before going into loop
        for j in range(len(rowWise)):      # Each line in temp, append with j'th row, i'th column
            temp.append(rowWise[j][i])
        colWise.append(temp)               # append colWise list with temp (essentially a column of data)
    return colWise
filename = 'raw_data.csv'        
lists = CSVdataRead(filename)

```
